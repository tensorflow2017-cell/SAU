```python

# DuckDB 연결 생성
con = duckdb.connect(database=':memory:', read_only=False)

# Sales 테이블 생성
print("--- 1. Sales 테이블 생성 (한글 필드명) ---")
con.sql("""
CREATE TABLE Sales (
    판매ID INTEGER PRIMARY KEY,
    지역 VARCHAR,
    상품 VARCHAR,
    수량 INTEGER,
    단가 INTEGER,
    매출액 INTEGER
);
""")

# Sales 테이블에 데이터 삽입
print("--- 2. Sales 테이블에 데이터 삽입 ---")
con.sql("""
INSERT INTO Sales (판매ID, 지역, 상품, 수량, 단가, 매출액) VALUES
(1, '서울', '노트북', 5, 1200000, 6000000),
(2, '부산', '태블릿', 10, 500000, 5000000),
(3, '서울', '모니터', 3, 300000, 900000),
(4, '제주', '노트북', 2, 1200000, 2400000),
(5, '부산', '노트북', 7, 1200000, 8400000),
(6, '서울', '태블릿', 8, 500000, 4000000),
(7, '제주', '모니터', 1, 300000, 300000),
(8, '서울', '노트북', 4, 1200000, 4800000);
""")

# 전체 데이터 확인
print("--- 3. Sales 테이블 데이터 확인 (SELECT) ---")
con.sql("SELECT * FROM Sales;").show()
```

```python
# 지역별 총매출액 (SUM) 및 판매 건수 (COUNT)
print("\n--- 4A. 지역별 총매출액 및 판매 건수 (GROUP BY 지역) ---")
con.sql("""
SELECT 
    지역, 
    SUM(매출액) AS 총매출,
    COUNT(판매ID) AS 총판매건수
FROM 
    Sales
GROUP BY 
    지역
ORDER BY 
    총매출 DESC;
""").show()
```

```python
# 지역 및 상품별 평균 판매 수량 (GROUP BY 지역, 상품)
print("\n--- 4B. 지역, 상품별 평균 판매 수량 (GROUP BY 지역, 상품) ---")
con.sql("""
SELECT 
    지역, 
    상품,
    AVG(수량) AS 평균수량
FROM 
    Sales
GROUP BY 
    지역, 상품
ORDER BY 
    지역, 상품;
""").show()

con.close()
```

```python
import duckdb
import os

# DuckDB 연결 생성 (인메모리 DB 사용)
con = duckdb.connect(database=':memory:', read_only=False)

# Sales_Data 테이블 생성
con.sql(f"""
CREATE TABLE Sales_Data (
    판매ID INTEGER PRIMARY KEY,
    지점명 VARCHAR,
    제품카테고리 VARCHAR,
    판매액 INTEGER,
    수량 INTEGER,
    할인율 DOUBLE
);
""")

print("--- 2. 판매 데이터 삽입 (총 12건) ---")
con.sql(f"""
INSERT INTO Sales_Data (판매ID, 지점명, 제품카테고리, 판매액, 수량, 할인율) VALUES
(1001, '강남', '가전', 500000, 1, 0.05),
(1002, '강남', '의류', 150000, 3, 0.10),
(1003, '강남', '가전', 800000, 1, 0.00),
(1004, '강남', '식품', 30000, 5, 0.05),
(1005, '종로', '식품', 70000, 2, 0.00),
(1006, '종로', '가전', 1200000, 1, 0.15),
(1007, '종로', '의류', 200000, 2, 0.10),
(1008, '종로', '의류', 180000, 1, 0.00),
(1009, '여의도', '식품', 45000, 4, 0.05),
(1010, '여의도', '가전', 400000, 1, 0.00),
(1011, '여의도', '의류', 90000, 1, 0.00),
(1012, '여의도', '의류', 250000, 2, 0.10);
""")

print("--- 3. Sales_Data 테이블 전체 데이터 확인 ---")
con.sql(f"SELECT * FROM Sales_Data;").show()
```

```python

# 4A. 지점별 주요 집계 (SUM, COUNT, AVG, MAX, MIN)
print("\n--- 4A. 지점별 주요 판매 지표 계산 (SUM, COUNT, AVG, MAX, MIN) ---")
con.sql(f"""
SELECT 
    지점명, 
    SUM(판매액) AS 총_판매액,
    COUNT(판매ID) AS 총_판매건수,
    AVG(판매액) AS 평균_판매액,
    MAX(판매액) AS 최고_단일_판매액,
    MIN(판매액) AS 최저_단일_판매액
FROM 
    Sales_Data
GROUP BY 
    지점명
ORDER BY 
    총_판매액 DESC;
""").show()

# 연결 종료
con.close()
""
```
